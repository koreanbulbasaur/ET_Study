1. 적절한 활성함수 사용

==> 모델 두개 사용  
==> 결과 예측은 ThoraricSurgery3.csv의 마지막 자료로 확인

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
```

```python
data_set = np.loadtxt('./data/ThoraricSurgery3.csv', delimiter=',')

X = data_set[:, :-1]    # 환자의 진찰 기록을 X로 지정
y = data_set[:, -1]     # 수술 1년 후 사망/생존 여부를 y로 지정함
```

## HIDDEN LAYER는 2개 층 -> 처음 node수가 30개, 두 번째 10개

### 모델 구조 결정

```python
model_sgd = Sequential()
model_sgd.add(Dense(30, input_dim=16, activation='relu'))
model_sgd.add(Dense(10, activation='relu'))
model_sgd.add(Dense(1, activation='sigmoid'))

model_adam = Sequential()
model_adam.add(Dense(30, input_dim=16, activation='relu'))
model_adam.add(Dense(10, activation='relu'))
model_adam.add(Dense(1, activation='sigmoid'))
```

### 모델 실행

```python
model_sgd.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'], loss_weights=None,
              sample_weight_mode=None, weighted_metrics=None)
model_adam.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'], loss_weights=None,
              sample_weight_mode=None, weighted_metrics=None)
history_sgd = model_sgd.fit(X, y, epochs=100, batch_size=16)
history_adam = model_adam.fit(X, y, epochs=100, batch_size=16)
```

```python
test_sample = data_set[-1, :-1]
test_sample = np.expand_dims(test_sample, axis=0)
predict_sgd = model_sgd.predict(test_sample)
predict_adam = model_adam.predict(test_sample)

print("SGD 모델 예측 결과:", predict_sgd)
print("Adam 모델 예측 결과:", predict_adam)
```

## 슬라이드 쇼

```python
import os
```

```python
img_files = os.listdir('./data/images/')
file_paths = []

for i in img_files:
    file_path = './data/images/' + i
    file_paths.append(file_path)

display(file_paths)
```

```python
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN,
                      cv2.WINDOW_FULLSCREEN)

for i in file_paths:
    img = cv2.imread(i, cv2.IMREAD_COLOR)

    if img is None:
        continue

    cv2.imshow('scene', img)

    if cv2.waitKey(1500) == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN,
                      cv2.WINDOW_FULLSCREEN)

idx = 0
while 1:
    img = cv2.imread(file_paths[idx], cv2.IMREAD_COLOR)

    if img is None:
        continue

    cv2.imshow('scene', img)

    if cv2.waitKey(1500) == ord('q'):
        break

    idx += 1

    if idx >= len(file_paths):
        idx = 0
cv2.destroyAllWindows()
```

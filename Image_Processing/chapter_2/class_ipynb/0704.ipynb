```python
import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt
```

```python
# 헥사 -> rgb
def hex_to_rgb(hex_code):
    hex_code = hex_code.lstrip('#').upper()
    length = len(hex_code)
    rgb = tuple(int(hex_code[i:i + length // 3], 16) for i in range(0, length, length // 3))
    return rgb
```

```python
img = np.ones((600, 1200, 3), np.uint8)*255

cv2.line(img, (50, 80), (200, 80), hex_to_rgb('#0A1647'), thickness=10, lineType=cv2.LINE_AA, shift=None)
cv2.line(img, (200, 80), (150, 300), hex_to_rgb('#8ffd5b'), 10, cv2.LINE_8)
cv2.arrowedLine(img, (300, 80), (300, 300), hex_to_rgb('#767BC3'), 10, cv2.LINE_AA)

cv2.rectangle(img, (50, 400), (300, 500), hex_to_rgb('#24e41e'), 10)

cv2.rectangle(img, (70, 420, 210, 60), hex_to_rgb('#59D6DD'), -1)

cv2.circle(img, (600, 200), 100, hex_to_rgb('#89a6f4'), 10, cv2.LINE_AA)

cv2.ellipse(img, (800, 300), (50, 100), 45, 0, 360, hex_to_rgb('#FC76fd'), 10, cv2.LINE_AA)

cv2.putText(img, 'Not Exist', (500, 500), cv2.FONT_HERSHEY_SIMPLEX, 1.5, hex_to_rgb('#cfa4fe'), 1, cv2.LINE_4)

cv2.imshow('img', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
black = (0, 0, 0)
thick = 30
line = cv2.LINE_AA
```

```python
import random

def random_rgb():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return r, g, b
```

```python
while 1:
    white_board = np.ones((600, 1200, 3), np.uint8)*255

    cv2.ellipse(white_board, (200, 200), (100, 65), 0, 0, 360, random_rgb(), thick, line)
    cv2.line(white_board, (150, 320), (150, 380), random_rgb(), thick, line)
    cv2.line(white_board, (250, 320), (250, 380), random_rgb(), thick, line)
    cv2.line(white_board, (60, 370), (60, 430), random_rgb(), thick, line)
    cv2.line(white_board, (60, 430), (330, 430), random_rgb(), thick, line)
    cv2.line(white_board, (50, 320), (350, 320), random_rgb(), thick, line)

    cv2.line(white_board, (430, 200), (630, 200), random_rgb(), thick, line)
    cv2.line(white_board, (630, 200), (530, 450), random_rgb(), thick, line)
    cv2.line(white_board, (700, 325), (780, 325), random_rgb(), thick, line)
    cv2.line(white_board, (700, 150), (700, 500), random_rgb(), thick, line)

    cv2.line(white_board, (900, 150), (1000, 150), random_rgb(), thick, line)
    cv2.line(white_board, (850, 200), (1050, 200), random_rgb(), thick, line)
    cv2.ellipse(white_board, (950, 300), (75, 50), 0, 0, 360, random_rgb(), thick, line)
    cv2.line(white_board, (1100, 250), (1050, 250), random_rgb(), thick, line)
    cv2.line(white_board, (1100, 350), (1050, 350), random_rgb(), thick, line)
    cv2.line(white_board, (850, 400), (850, 450), random_rgb(), thick, line)
    cv2.line(white_board, (850, 450), (1050, 450), random_rgb(), thick, line)
    cv2.line(white_board, (1100, 150), (1100, 450), random_rgb(), thick, line)

    cv2.imwrite('name.jpg', white_board)

    cv2.imshow('name', white_board)

    if cv2.waitKey() == ord('q'):
        break
    else:
        continue

cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('Video open failed')
    sys.exit()

while 1:
    ret, frame = cap.read()

    if not ret:
        print('frame read failed')
        break

    frame_flipped = cv2.flip(frame, 1)


    cv2.circle(frame, (100, 100), 50, random_rgb(), 5, line)
    edge = cv2.Canny(frame, 50, 150)

    cv2.imshow('image', frame)
    cv2.imshow('edge', edge)
    cv2.imshow('image', frame_flipped)


    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('Video open failed')
    sys.exit()

h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('output.avi', fourcc, fps, (w, h), isColor=False)

while 1:
    ret, frame = cap.read()

    if not ret:
        print('frame read failed')
        break

    # cv2.circle(frame, (200, 100), 100, random_rgb(), thick, line)

    frame = cv2.flip(frame, 1) # 좌우 반전

    edge = cv2.Canny(frame, 50, 100)

    cv2.imshow('frame', frame)
    cv2.imshow('edge', edge)

    out.write(edge)

    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
out.release()
```

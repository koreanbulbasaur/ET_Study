```python
import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('./data/cow.png')
dst = cv2.imread('./data/green.png')

cv2.imshow('src', src)
cv2.imshow('dst', dst)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

_, mask = cv2.threshold(src_gray, 240, 255, cv2.THRESH_BINARY_INV)

h, w = src.shape[:2]

dst_crop = dst[500:500+h, 300:300+w]
cv2.copyTo(src, mask, dst_crop)

cv2.namedWindow('dst', cv2.WINDOW_NORMAL)
cv2.imshow('dst', dst)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import os
```

```python
img_paths = os.listdir('../chapter_1/data/images/')

print(img_paths)
```

```python
file_names = []
for img_file in img_paths:
    file_name = '../chapter_1/data/images/' + img_file
    file_names.append(file_name)

cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

idx = 0
while 1:
    img = cv2.imread(file_names[idx])
    if img is None:
        continue

    cv2.imshow('scene', img)

    if cv2.waitKey(3000) == ord('q'):
        break

    idx += 1

    if idx >= len(file_names):
        idx = 0

cv2.destroyAllWindows()
```

```python
img_BGR = cv2.imread('./data/cat.bmp')
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)
img_GRAY = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2GRAY)

fig, ax = plt.subplots(1, 3, figsize=(12, 6))

ax[0].imshow(img_BGR)
ax[1].imshow(img_RGB)
ax[2].imshow(img_GRAY, cmap='gray')

plt.show()
```

```python
img = cv2.imread('./data/imgbin_hat.png', cv2.IMREAD_UNCHANGED)

h, w = img.shape[:2]

img = cv2.resize(img, (w//4, h//4), interpolation=cv2.INTER_AREA)

hat_img = img[:, :, 0:-1]
hat_mask = img[:, :, -1]
cv2.imshow('hat', hat_img)
cv2.imshow('mask', hat_mask)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

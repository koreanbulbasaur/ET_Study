```python
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('../data/lenna.bmp', 1)

# dst = np.clip(src + 100., 0, 255).astype(np.uint8)
dst = cv2.add(src, (100, 100, 100, 0))

cv2.imshow('src', src)
cv2.imshow('dst', dst)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
src = cv2.imread('../data/lenna256.bmp', 0)
src_2 = np.zeros_like(src)
cv2.circle(src_2, (128, 128), 100, 200, -1)
cv2.circle(src_2, (128, 128), 50, 50, -1)

dst_1 = cv2.add(src, src_2)
dst_2 = cv2.addWeighted(src, 0.5, src_2, 0.5, 0)
dst_3 = cv2.subtract(src, src_2)
dst_4 = cv2.absdiff(src, src_2)

fig, ax = plt.subplots(2, 3, figsize = (12, 4))

lst = [src, src_2, dst_1, dst_2, dst_3, dst_4]
idx = 0

for x in range(2):
    for y in range(3):
        ax[x, y].imshow(lst[idx], cmap='gray')
        ax[x, y].axis('off')
        idx += 1

# cv2.imshow('src', src)
# cv2.imshow('src_2', src_2)
# cv2.imshow('dst_1', dst_1)
# cv2.imshow('dst_2', dst_2)
# cv2.imshow('dst_3', dst_3)
# cv2.imshow('dst_4', dst_4)

# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

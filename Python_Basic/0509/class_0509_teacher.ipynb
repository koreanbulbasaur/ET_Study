```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.font_manager as fm 
from matplotlib import rc
import matplotlib as mpl
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16] # y = x**2
plt.figure(figsize = (4, 3))
plt.plot(x, y)
plt.show()
```

```python
data_dict = {
    'data_x':[1, 2, 3, 4],
    'data_y': [1, 4, 9, 16]
}
plt.figure(figsize = (3, 3))
plt.plot('data_x', 'data_y', data = data_dict)
x_fonts = {'family':'monospace',
                       'color': 'r',
                       'fontsize': 20,
                       'weight': 'bold'}

plt.xlabel('X', labelpad = 10, color = 'b', 
           family = 'monospace', 
           size = 20, weight = 'bold')
# plt.xlabel('X', labelpad = 10, 
#            fontdict = x_fonts)
plt.ylabel('X^2')
plt.title('X vs X^2')

plt.show()
```

```python
plt.figure(figsize = (3, 3))
plt.plot('data_x', 'data_y', data = data_dict,
         label = 'Price ($)')

##{'left', 'center', 'right'}
plt.xlabel('X', loc = 'center')

##{'bottom', 'center', 'top'}
plt.ylabel('Y', loc = 'center')

plt.legend(loc = 'best')

plt.show()
```

```python
plt.figure(figsize = (4,3))
plt.plot(np.arange(1, 20), np.arange(1, 20)**3, label = 'X^3')
plt.plot(np.arange(1, 20), np.arange(1, 20)**2, label = 'X^2')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
plt.legend(ncol = 2, fontsize = 8, frameon = True,
           edgecolor = 'r')
plt.show()
```

```python
plt.figure(figsize = (4,3))
plt.plot(np.arange(1, 20), np.arange(1, 20)**3, label = 'X^3')
plt.plot(np.arange(1, 20), np.arange(1, 20)**2, label = 'X^2')

plt.xlabel('X')
plt.ylabel('Y')
# plt.xlim(0, 10)
# plt.ylim(0, 1000)
# [xmin, xmax, ymin, ymax]
plt.axis([0, 10, 0, 1000])
plt.title('Title')
plt.legend()

plt.show()
```

```python
## line type
plt.figure(figsize = (4,3))
plt.plot(np.arange(1, 20), np.arange(1, 20)**3, label = 'Dashed',
         linestyle = '--')
plt.plot(np.arange(1, 20), np.arange(1, 20)**2, label = 'Dotted',
         linestyle = ':')

plt.xlabel('X')
plt.ylabel('Y')
plt.axis([0, 10, 0, 1000])
plt.title('Title')
plt.legend()

plt.show()
```

```python
plt.figure(figsize = (6, 4))
plt.plot([1, 2, 3], [4, 4, 4], linestyle = '-', marker= 'o', 
         color = 'r', label = 'Solid', linewidth = 2)
plt.plot([1, 2, 3], [3,3, 3], linestyle = '--', marker= '^',
         color = 'b', label = 'Dashed')
plt.plot([1, 2, 3], [2, 2, 2], linestyle = ':', marker= '+',
         color = 'c', label = 'Dotted',  linewidth = 2)
plt.plot([1, 2, 3], [1, 1, 1], linestyle = '-.', 
         color = 'm', label = 'Dash-dotted')
plt.ylim(0, 5)
plt.legend(ncol = 2, loc = 'upper center')
plt.show()
```

```python
t = np.array([1, 2, 3, 4])


plt.plot(t, t, 'ro')
plt.plot(t, t**2, 'g^--')
plt.plot(t, t**3, 'bx-.')


plt.show()
```

```python
x = np.arange(0, 2, 0.2)
# print(x)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, 'r^--', markersize = 2, linewidth = 2)

plt.xlabel('Months')
plt.ylabel('Sales')

plt.xticks([0, 1, 2], labels = ['Jan', 'Feb', 'March'])
plt.yticks(np.arange(0, 4), 
           labels = ['100$', '200$', '300$', "400$"])

plt.tick_params(axis = 'x', direction = 'in', 
                   length = 3, pad = 6, labelsize = 14, 
                   labelcolor = 'm', rotation = 45)

plt.show()
```

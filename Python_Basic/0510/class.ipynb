```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(3)
y_value = [100, 400, 600]
y_std = [20, 43, 22]
plt.barh(x, y_value, color=['m', 'r', 'b'], alpha = 0.5,
         height=0.6) 
plt.title('Title')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(x, labels=['2019', '2020', '2021'])

plt.show()
```

```python
x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y)
plt.show()
```

```python
np.corrcoef(x, y)
```

```python
x = np.random.randn(100)
y_cor = x**2
y = np.random.randn(100)

plt.scatter(x, y_cor)
plt.show()
```

```python
x = np.random.normal(10, 1, 100)
y_cor = x + np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y_cor)
plt.show()
```

```python
x = np.random.normal(10, 1, 100)
y_cor = x + np.random.randn(100)
y_size = np.random.normal(100, 100, 100)
y = np.random.randn(100)

plt.scatter(x, y_cor, s = y_size, alpha=0.5)
plt.show()
```

```python
x = np.random.randn(100)

plt.hist(x, bins=30)
plt.show()
```

```python
x = np.random.randn(10000)
y = np.random.uniform(-3, 3, 10000)

plt.hist(x, bins=30, label='Nornal dist', alpha = 0.5)
plt.hist(y, bins=30, label='Uniform dist', alpha = 0.5)
plt.legend()
plt.show()
```

```python
a = np.random.normal(1, 2, 100)
b = np.random.normal(3, 2, 100)
c = np.random.normal(5, 2, 100)

plt.boxplot([a, b, c])

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dict = {
    'x' : np.arange(1, 5),
    'y' : np.arange(2, 6)
}

plt.plot('x', 'y', data = data_dict,
         linestyle='--', marker = 'o',
         color='r', label='solid')
plt.title('Title')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
data_dict = {
    'x' : np.arange(1, 5),
    'y' : np.arange(2, 6)
}

plt.plot('x', 'y', data = data_dict,
         linestyle='--', marker = 'o',
         color='r', label='solid')
plt.title('Title')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(alpha= 0.3)
plt.show()
```

```python
data_dict = {
    'x' : np.arange(1, 5),
    'y' : np.arange(2, 6)
}

x_font = {
    'family':'monospace',
    'fontsize' : 15,
    'color' : 'r',
    'weight' : 'bold'
}

plt.plot('x', 'y', data = data_dict,
         linestyle='--', marker = 'o',
         color='r', label='solid')
plt.title('Title')
plt.xlabel('X', fontdict=x_font, labelpad=20)
plt.ylabel('Y')
plt.legend()
plt.grid(alpha= 0.3)
plt.show()
```

```python
x = np.arange(1, 5)

plt.plot(x, x**2, label='x^2', linestyle='-')
plt.plot(x, x**3, label='x^3')
plt.xlim(1, 3)
plt.ylim(0, 30)
plt.xticks([1, 2, 3], labels=['Jan', 'Feb', 'Mar'])
```

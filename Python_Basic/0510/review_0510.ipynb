```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dict = {
    'x': np.arange(1, 5),
    'y': np.arange(2, 6)
}

x_fonts = {
    'family':'monospace',
    'fontsize': 15,
    'color': 'r',
    'weight': 'bold'
}

plt.plot('x', 'y', data = data_dict,
         linestyle = '--', marker = 'o',
         color = 'r', label = 'Solid')
plt.title('Title')

plt.xlabel('X', fontdict=x_fonts, labelpad= 20)
plt.ylabel('Y')
plt.legend()
plt.grid(alpha = 0.3, linestyle = '--')
plt.show()
```

```python
x = np.arange(1, 5)

plt.plot(x, x**2, marker = 'o', linestyle = '-', label = 'x^2')
plt.plot(x, x**3, marker = '^', linestyle = '--',label = 'x^3')
plt.xlim(1, 3)
plt.ylim(0, 30)
plt.xticks([1, 2, 3], labels = ['Jan', 'Feb', 'Mar'],
           rotation = 90)
plt.tick_params(axis = 'x', direction = 'in',
                length = 2, pad = 6)

plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (12, 10))

x = np.arange(1, 3, 0.1)

ax[0, 0].plot(x, x, 'ro-', label = 'linear')
ax[0, 1].plot(x, x**2, 'b*--', label = 'square')
ax[1, 0].plot(x, x**3, 'g^:', label = 'cubic')
ax[1, 1].plot(x, x**4, 'mD-.', label = 'quadratic')

ax[0, 0].legend()
ax[0, 1].legend()
ax[1, 0].legend()
ax[1, 1].legend()

# plt.legend()
plt.show()

```

```python
## 

# x = np.arange(3)
y = np.arange(3)

# y_value = [100, 400, 600]
x_value = [100, 400, 600]

y_std = [20, 34, 22]
plt.barh(y, x_value, color = ['m', 'r', 'b'] , alpha = 0.5,
        height = 0.6, xerr = y_std)

plt.title('Title')
plt.xlabel('Year')
plt.ylabel('Value')
plt.yticks(y, labels=['2019', '2020', '2021'])

plt.show()
```

```python
##  

x = np.random.randn(1000)
y = np.random.uniform(-3, 3, 1000)

plt.hist(x, bins= 30, label = 'Normal dist', alpha = 0.5)
plt.hist(y, bins= 30, label = 'Uniform dist', alpha = 0.5)
plt.legend()
plt.show()
```

```python
## Boxplot

a = np.random.normal(1, 2, 100)
b = np.random.normal(5, 4, 100)
c = np.random.normal(3, 2, 100)

plt.boxplot([a, b, c], whis = 2.5, notch=True, vert=True,
            showmeans=True)

plt.show()
```

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
# titanic.dtypes
```

```python
fig, ax = plt.subplots(1, 2)

# ax[0].scatter(titanic['age'], titanic['fare'])
# ax[1].scatter(titanic['age'], titanic['fare'])

sns.regplot(x = 'age', y = 'fare', data = titanic,
            marker = '+', 
            ax = ax[0], fit_reg = False)
sns.regplot(x = 'age', y = 'fare', data = titanic,
            ax = ax[1], fit_reg = True)

plt.show()
```

```python
## distplot
# darkgrid, whitegrid, dark, white, ticks
sns.set_style('darkgrid')

fig, ax = plt.subplots(1, 3, figsize=(12, 5))

sns.distplot(titanic['fare'], ax = ax[0])
sns.distplot(titanic['fare'], hist = False, ax = ax[1])
sns.distplot(titanic['fare'], kde = False, ax = ax[2])

ax[0].set_title('dist + kde')
ax[1].set_title('kde')
ax[2].set_title('dist')


plt.show()
```

```python
## heatmap
# titanic.head()

table = titanic.pivot_table(index = ['sex'], columns=['class'], aggfunc='size')
# table.sum(axis = 1)
sns.heatmap(table, annot=True)
plt.show()
```

```python
## strip, swarm

fig, ax = plt.subplots(1, 2, figsize = (12, 6))

sns.stripplot(x = 'class', y = 'age', data = titanic,
              ax = ax[0])
sns.swarmplot(x = 'class', y = 'age', data = titanic,
              ax = ax[1])

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 6))

sns.barplot(x = 'sex', y = 'survived', data = titanic,
            ax = ax[0])
sns.barplot(x = 'sex', y = 'survived', data = titanic,
            hue = 'embarked',
            ax = ax[1])

plt.show()
```

```python
##

fig, ax = plt.subplots(1, 3, figsize = (12, 6))

sns.boxplot(x = 'alive', y = 'age', data = titanic,
            ax = ax[0])
sns.boxplot(x = 'alive', y = 'age', hue = 'sex',
            data = titanic,
            ax = ax[1])
sns.violinplot(x = 'alive', y = 'age', hue = 'sex',
            data = titanic,
            ax = ax[2])

plt.show()
```
